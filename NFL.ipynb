```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

#Getting Passing Stats

```python
url = 'https://www.nfl.com/stats/team-stats/'
page = requests.get(url)
soup = BeautifulSoup(page.text, 'html')

#gettign passing stats from nfl .com
passing = soup.find_all('table')[0]
table = passing.find_all('th')
table = [tab.text.strip() for tab in table]
table

#renaming columns
passing_df = pd.DataFrame(columns=table)
passing_df.rename(columns={'Rate': 'QBR', '1st': '1st_downs','1st%': '1st_down_percentage', '1st': '1st_downs', 'Sck': 'Scks_against' }, inplace=True)

passing_df
```

```python
stats = soup.find_all('td')
stats = [stat.text.strip() for stat in stats]

cleaned_team_names = [name.split('\n')[0].strip() for name in stats]

names = np.reshape(cleaned_team_names, (32, 16))
names = pd.DataFrame(names)
names.columns = passing_df.columns
passing_df = names
```

#Cleaning the passing data

```python
passing_df= passing_df.drop(['20+','40+', 'Lng', 'SckY'],axis=1)

```

```python
cols = ['Att',	'Cmp',	'Cmp %'	,'Yds/Att',	'Pass Yds',	'TD',	'INT',	'QBR',	'1st_downs',	'1st_down_percentage',	'Scks_against']
for c in cols:
  passing_df[c] = passing_df[c].astype(float)
```

#Getting Rushing Data

```python
#creating russhing df
url = 'https://www.nfl.com/stats/team-stats/offense/rushing/2024/reg/all'
page = requests.get(url)
soup = BeautifulSoup(page.text, 'html')

# #second table in the wiki page
rushing = soup.find_all('table')[0]
table = rushing.find_all('th')
table = [tab.text.strip() for tab in table]
table

rushing_df = pd.DataFrame(columns=table)
rushing_df.rename(columns={'Rate': 'QBR', '1st': '1st Downs','1st%': '1st Down Percentage', '1st': '1st_downs' }, inplace=True)
# pasing_df.remove('SckY')
rushing_df
```

```python
#populating rushing df
stats = soup.find_all('td')
stats = [stat.text.strip() for stat in stats]

cleaned_team_names = [name.split('\n')[0].strip() for name in stats]
rush = np.reshape(cleaned_team_names, (32, 11))
rush = pd.DataFrame(rush)
rush.columns = rushing_df.columns
rushing_df = rush
```

#Cleaning Rushing Data

```python
rushing_df= rushing_df.drop(['20+','40+', 'Lng'],axis=1)
```

```python
cols = ['Att',	'Rush Yds',	'YPC'	,	'TD',	'Rush 1st' ,'Rush 1st%' ,'Rush FUM']

for c in cols:
  rushing_df[c] = rushing_df[c].astype(float)
```

```python
cols = ['Rec',	'Yds'	,'Yds/Rec'	,'TD'	,'Rec 1st'	,'Rec 1st%',	'Rec FUM']
for c in cols:
  reciving_df[c] = reciving_df[c].astype(float)
```

#Get Team Stats

```python
url = 'https://www.nfl.com/standings/league/2024/REG'
page = requests.get(url)
soup = BeautifulSoup(page.text, 'html')

reciving = soup.findAll('table')[0]
table = reciving.findAll('th')
table = [tab.text.strip() for tab in table]

table
team_df = pd.DataFrame(columns=table)
```

```python
team_stats = soup.find_all('td')
team_stats = [stat.text.strip() for stat in team_stats]

cleaned_data = [name.split('\n')[0].strip() for name in team_stats]

cleaned_data
team = np.reshape(cleaned_data, (32, 17))
team = pd.DataFrame(team)
team.columns = team_df.columns
team_df= team
```

#Clean team data

```python
team_df=team_df.rename(columns={'NFL Team':'Team'})
team_df= team_df.drop(['Div',	'Pct',	'Conf',	'Pct',	'Non-Conf'	,'Strk',	'Last 5'],axis=1)
team_df['Team'] = team_df['Team'].apply(lambda x: x.rsplit(" ", 1)[-1])
```

```python
team_df[['Home wins', 'Home losses', 'Home ties']] = team_df['Home'].str.split('-', expand=True)
team_df[['Road wins', 'Road losses', 'Road ties']] = team_df['Road'].str.split('-', expand=True)
```

```python
col = [	'W',	'L',	'T',	'PCT',	'PF',	'PA',	'Net Pts', 'Home wins', 'Home losses', 'Home ties','Road wins', 'Road losses', 'Road ties']
for c in col:
  team_df[c] = team_df[c].astype(float)
```

```python
team_df= team_df.drop(['Home',	'Road'],axis=1)
```

```python
passing_df=passing_df.sort_values('Team')
rushing_df=rushing_df.sort_values('Team')
reciving_df=reciving_df.sort_values('Team')
team_df=team_df.sort_values('Team')
```

```python
# Merge passing_df and rushing_df on 'Team'
pass_rec = pd.merge(passing_df, reciving_df, on='Team', how='inner')
pass_rec_rush = pd.merge(pass_rec, rushing_df, on="Team", how= "inner")

offence = pd.merge( team_df,pass_rec_rush, on="Team", how="inner")
offence
```
